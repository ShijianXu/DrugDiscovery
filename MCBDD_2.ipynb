### Retrieve all approved drugs from the ChEMBL database, sort them by approval year and name

```python
molecule = new_client.molecule
print(len(molecule))
```

```python
# approved drugs
approved_drugs = molecule.filter(max_phase=4)
print(len(approved_drugs))
```

```python
# store approved drugs in a dataframe
approved_drugs_df = pd.DataFrame(approved_drugs)
```

```python
approved_drugs_df.head()
```

```python
print(approved_drugs_df.columns)
```

```python
# sort by 'first_approval' and 'pref_name'
approved_drugs_df = approved_drugs_df.sort_values(['first_approval', 'pref_name'])
approved_drugs_df.head()
```

### Get all the drugs approved since 2019

```python
drug_2019_df = approved_drugs_df[approved_drugs_df['first_approval'] >= 2019]
print(len(drug_2019_df))
```

```python
drug_2019_df.head()
```

```python
mechanism = new_client.mechanism

# Prepare list to collect (drug, target_chembl_id, uniprot_id)
targets = set()

for chembl_id in tqdm(drug_2019_df['molecule_chembl_id']):
    try:
        # Get mechanisms of action
        mechanisms = mechanism.filter(molecule_chembl_id=chembl_id)
        for mech in mechanisms:
            targets.add(mech['target_chembl_id'])

    except Exception as e:
        print(f"Error processing {chembl_id}: {e}")

# Convert set to list
targets = list(targets)
print(len(targets))
print(targets)
```

```python
target = new_client.target

uniprot_ids = []
for targ in tqdm(targets):
    try:
        # Get target information
        target_info = target.get(targ)
        for comp in target_info['target_components']:
            for xref in comp['target_component_xrefs']:
                if xref['xref_src_db'] == 'UniProt':
                    uniprot_id = xref['xref_id']
                    uniprot_ids.append((targ, uniprot_id))    

    except Exception as e:
        print(f"Error processing {targ}: {e}")

print(len(uniprot_ids))
```

```python
print(uniprot_ids)
```

### retrieve UniProt keywords

```python
import uniprot as uni
```

```python
import requests

protein_info = []

def get_uniprot_entry(uniprot_id, format="json"):
    url = f"https://rest.uniprot.org/uniprotkb/{uniprot_id}.{format}"
    response = requests.get(url)
    if response.ok:
        return response.json()
    else:
        print(f"Failed to retrieve {uniprot_id}: {response.status_code}")
        return None
```

```python
protein_keywords = {}
for targ, uniprot_id in tqdm(uniprot_ids):
    entry = get_uniprot_entry(uniprot_id)
    if entry:
        try:
            protein_keywords[uniprot_id] = entry['keywords']
        except KeyError:
            print(f"Keywords not found for {uniprot_id}")
            protein_keywords[uniprot_id] = None
```

```python
for uniprot_id, keywords in protein_keywords.items():
    print(f"UniProt ID: {uniprot_id}, Keywords: {keywords}")
```
